```python
sql = """
SELECT * FROM (
    SELECT MOVIENM, MAX_AUDIACC
    FROM MOVIE2 
    WHERE OPENDT < TO_DATE(:opendate, 'YY/mm/dd')
    ORDER BY MAX_AUDIACC DESC
) WHERE rownum <= 20
"""
```

```python
for y in range(3, 20):
    if y < 10:
        opendate = '0'+str(y)+'/01/01'
        year = '200'+str(y)
    else:
        opendate = str(y)+'/01/01'
        year = '20'+str(y)
        
    params = {
        'opendate' : opendate
    }
    
    result_data = db.execute(sql, params)
    
#     print(year, result_data)
    for i in result_data:
        movie_list.append(i[0])

movie_list = list(set(movie_list))
print(movie_list)
```
