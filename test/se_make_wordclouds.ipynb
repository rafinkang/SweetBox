```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from DbConn import *
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
def make_wordclouds():
    db = DbConn()
    
    select_query = """
    select moviecd, prodnm
    from movie2  
    order by prodcd
    """
    
    select_result = pd.read_sql(select_query,db.connection)

    print("제작사 테이블 가져오기 -----")
    db.disconnect()
    return  select_result
```

```python
df = make_wordclouds()
words=df['PRODNM'].value_counts()
```

```python
img_mask=np.array(Image.open('./map.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf',mask=img_mask,
                      stopwords=stopwords,
                      background_color='white',colormap = "Accent_r",
                      width=2000, height=2000).generate_from_frequencies(words) 
plt.figure(figsize=(10,10))
plt.imshow(wordcloud,interpolation='bilinear')
plt.show()

```

```python
img_mask=np.array(Image.open('./worldmap.png'))
stopwords = set(STOPWORDS)
stopwords.add("said")
wordcloud = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf',mask=img_mask,
                      stopwords=stopwords,
                      background_color='black',colormap = "Accent_r",
                      width=2000, height=2000).generate_from_frequencies(words) 
plt.figure(figsize=(10,10))
plt.imshow(wordcloud,interpolation='bilinear')
plt.show()

```
